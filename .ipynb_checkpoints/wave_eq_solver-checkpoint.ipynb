The following file presents a simple 1D wave equation solver.

The wave equation reads
$$ \frac{\partial^2{}}{\partial{t^2}}u(x,t) = c^2\frac{\partial^2{}}{\partial{x^2}}u(x,t)~~.$$

In the case of acoustics $u ({x,t})$ represents the pressure field in the medium. The equation is derived by utilizing the conservation of mass (continua eq.) and momentum (Euler's eq.), along with the state equation for an ideal gas. Linearizing near a stationary solution (assuming small pertubations) leads to the wave equation. 

The equation is solved by casting in two a set of first order differential equations
$$\frac{\partial{}}{\partial t}\left[{\begin{array}{c}
u\\
v
\end{array}}\right]=\left[{\begin{array}{cc}
0 & 1\\
c^{2}\frac{\partial{^{2}}}{{\partial x^{2}}} & 0
\end{array}}\right]\left[{\begin{array}{c}
u\\
v
\end{array}}\right]~~. $$
The various solution methods correspond to different ways by which to linear porpagator is approximated.


The file is contructed in an increasing order of complexity. We begin with a simple model and increase the complexity step by step.
## Steps:

1. Finite difference solution
2. Solution employing FFT algorithm - an accurate differentiation
3. Solution employing Chebychev polynomials
4. Applying a source - mapping a rapid pulse in the boundary to an initial condition.
5. Spatial dependent speed field
6. Measurements

## Initialization

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from derivative_with_fft import deriv_n_gen


# Parameters
L = 1.0         # Length of the domain
Nx = 2**9        # Number of spatial points
c = 0.01         # Wave speed
x = np.linspace(0, L, Nx)
dx = x[1] - x[0]
```

## Initial conditions
We study to initial conditions (i) eigenvector with constant velocity (ii) gaussian with zero velocity

```python
## Eigenvector with constant velocity
k = 2*np.pi/L
omega = c*k

def initial_displacement(x,k):
    return np.sin(k*x)  

def initial_velocity(x,k):
    return -c*k*np.cos(k*x)

## Gaussian with initial zero velocity
def initial_gaussian(x,mu,sig):
    N = 1/np.sqrt(2*np.pi*sig**2)
    return N*np.exp(-(x-mu)**2/(2*sig**2))

def initial_gaussian_velocity(x,mu,sig):
    return np.zeros(x.shape)
```

```python
## Eigenvector initial state with constant velocity
u0 = initial_displacement(x,k)
v0 = initial_velocity(x,k)
y0 = np.concatenate([u0, v0])
```

```python
## Gaussian initial state with zero velocity
mu = 1/2
sig = L/10
u0 = initial_gaussian(x, mu, sig)
v0 = initial_gaussian_velocity(x, mu, sig)
y0 = np.concatenate([u0, v0])
```

## Propagator
Two methods of solution are presented:
1. Representing the linear operator in terms of a finite difference equation
2. Utilizing FFT to evaluate the partial derivative accurately

### Finite-difference

```python
# Second spatial derivative using central differences
def laplacian(u, dx):
    '''Finite difference with periodic boundary conditions '''
    d2u = np.zeros_like(u)
    d2u[1:-1] = (u[2:] - 2*u[1:-1] + u[:-2]) / dx**2
    d2u[0] = (u[1] - 2*u[0] + u[-1]) / dx**2
    d2u[-1] = (u[0] - 2*u[-1] + u[-2]) / dx**2
    return d2u

# ODE system: y = [u, v]
def wave_eq(t, y):
    u = y[:Nx]
    v = y[Nx:]
    du_dt = v
    dv_dt = c**2 * laplacian(u, dx)
    return np.concatenate([du_dt, dv_dt])
```

### FFT solution

```python
# Second spatial derivative using fft
def laplacian(u, x):
    '''Finite difference with periodic boundary conditions '''
    return deriv_n_gen(u,x,2)


# ODE system: y = [u, v]
def wave_eq(t, y):
    u = y[:Nx]
    v = y[Nx:]
    du_dt = v
    dv_dt = c**2 * laplacian(u, x)
    return np.concatenate([du_dt, dv_dt])
```

## Propagation
Propagation utilizing scipy Runge Kutta 45

```python
# Time span
k0 = np.pi/L
omega0 = c*k0;
T0 = L/(10*c)
t_span = (0, T0)
t_eval = np.linspace(0,t_span[1],30)
# Propagate
sol = solve_ivp(wave_eq, t_span, y0, t_eval=t_eval, method='RK45')
```

## Chebychev solution

Next we utilize expand the dynamical propagator in a Chebychev series:
$$\vec{u}(dt) = \exp(O dt) [\vec{u}(0)] = \sum_n d_n(\lambda_{\text{max}},\lambda_{\text{min}},dt)\phi_n(\bar{O})[\vec{u}(0)]~~.$$
Here, $O$ is a normal linear differential operator with maximum/minimum eigenvalues $\lambda_{\text{max}}/\lambda_{\text{min}}$, $\bar{O}$ is normalized version of $O$ with eigenvalues in the range $[-1,1]$, $dt$ is the time step, $\vec{u}(0)$ is the initial state,
$$d_n = \exp((\lambda_{\text{min}}+\lambda_{\text{max}}) dt/2)c_n~~,$$
with $c_n = 2 I_{n}(R)$ for $n>0$ and $c_0 = I_0(R)$, where $R = (\lambda_{\text{max}} - \lambda_{\text{min}})dt/2$. The differential operators $\phi_n$ satisfy the recursion relation of the Chebychev polynomials
$$\phi_{n+1} = 2 O\phi_{n} - \phi_{n-1}~~. $$
Such a series provides an approximation that is nearly as good as the best possible polynomial approximation in the sense of minimizing the maximum error. As a result, the expansion allows an extremly efficient method by which to propogate the initial state.

The code is based on the paper [Propagation Methods for Quantum Molecular Dynamics - R. Kosloff](https://scholars.huji.ac.il/sites/default/files/ronniekosloff/files/annurev.pc_.45.100194.pdf).

```python
## Defining functions

import scipy

def besseli(v,z):
    '''
    Modified Bessel function of the first kind of real order.
    Input:
        v: float real, order of the function
        z: float or complex, argument   
    '''
    return scipy.special.iv(v,z)

def cheby_coeff(dt,lam_max,lam_min):
    '''
    Evaluates the Chebychev coefficients for the propagator exp(O*dt),
    where O is a linear differential operator with units of 1/time.
    
    Based on the paper: Propagation Methods for Quantum Molecular Dynamics - Ronnie Kosloff 1994

    Input:
        dt: float, the time interval of propagation.
        lam_min: float, expected minimum eigenvalue of O
        lam_max: float, expected max eigenvalue of O
    
    Note: lam_min and lam_max do not have to be exact. Although it is important 
          to keep the normalized range of eigenvalues within [-1,1].
    
    Retun: array, contains the Chebychev coefficients, which can be used to propagate the system.
    
    '''
    # Factor to ajust the eigenvalues to be between [-1,1]
    dE = lam_max-lam_min  #Note that dE has units of 1/time
    R = dE*dt/2
    Nmax = 1000        # Maximum number of coefficients 
    c_j = np.zeros(Nmax,dtype = complex)
    c_j[0] = besseli(0,R)           #Zero coefficient.
    # The vector a contans c_n coefficients from 1 to Nmax, further multipication by exp(lam_min+ *dt) is needed  
    for n in range(1,Nmax):
        c_j[n] = 2*besseli(n,R)
        if abs(c_j[n])<1e-17 and n>R: 
            break

    Nmax = n+1
    d_j = np.exp((lam_min+lam_max)*dt/2)*c_j #normalizing factor times the cooefficeints
    return d_j,Nmax


def propagator(vec,d_j,Nmax,O,Nt,tup):
    """
    Evaluates exp(O*t)*vec, with t=Nt*dt, utilizing a Chebychev expansion of the dynamical propagator.

    Input: 
        vec: numpy array, containing the initial state
        dj: numpy array, the Chebychev coefficients
        O: function, the dynamical generator (e.g, a differential operator)
        tup: tuple, containing the dynamical generator additional parameters
    """
    fi = np.zeros((vec.shape[0],3),dtype = complex)
    #x,max_x,min_x,lx,mass,k,lam_min,lam_max,dk = tup
    for j in range(Nt):     # Running over the time steps
        fi[:,0] = vec
        # The normalized differential operator O
        fi[:,1] = O(vec,tup)              
        G1 = d_j[0]*fi[:,0]
        G2 = d_j[1]*fi[:,1]
        G_3 = G1+G2
        for i in range(1,Nmax-1):
            fi[:,2] = 2*O(fi[:,1],tup)-fi[:,0]
            fi[:,0]=fi[:,1]
            fi[:,1]=fi[:,2]
            G_3 = G_3+d_j[i+1]*fi[:,2]
        #norm[j] = sum((np.abs(G_3))**2)*dx;
        vec = G_3
    return vec

def O(y,tup):
    '''
    Normalized wave equation dynamical generator
        Input:
            y: array, the initial state 
            tup: tuple, containing the following variables
            x_max: float, max(x), where x is the grid
            x_min: float, min(x)
            lx: int, length of x
            lam_min: float, minimum eigenvalue
            lam_max: float, maximum eigenvalue
            dk: float, the spacing in k-space
            c: array, speed of sound
        
        Return: Normalized version of the wave equation's dynamical generator
    '''
    lam_min,lam_max,dk,c = tup
    u = y[:Nx]
    v = y[Nx:]
    du_dt = v
    dv_dt = c**2 * deriv_n_gen(u,x,2)
    dE =lam_max-lam_min
    df = np.concatenate([du_dt, dv_dt])
    # The normalized operator, in the appropriat form for the Chebychev propagator.
    O = (2/dE)*(df-lam_min*y)-y
    return O
```

```python
# Constants
pi = np.pi

# Model constants paramenter definition    
dx = L/Nx
kmax = pi/dx
dk = 2*pi/(Nx*dx)

## Chebychev coefficents decleration
dt = 0.5
Nt = int(T0//dt)
t = dt*Nt
lam_max = np.sqrt((c**2)*kmax**2+1) 
lam_min = - lam_max
R = (lam_max-lam_min)*dt/2
d_j,Nmax = cheby_coeff(dt,lam_max,lam_min)  

## Declarations for the time propagation
lx = len(x)
max_x = max(x)
min_x = min(x)
```

```python
## Propagation in time
tup = (lam_min,lam_max,dk,c)
y_t_cheby = propagator(y0,d_j,Nmax,O,Nt,tup)
u_t_cheby = y_t_cheby[:Nx]
```

## Analytical solution

```python
## Eigenvector
def ev_analyt(t):
    return np.sin(k*x-omega*t)

## Gaussian
def gauss_t(x,mu,sig,c,t):
    N = 1/np.sqrt(2*np.pi*sig**2)
    return N*np.exp(-(x-mu-c*t)**2/(2*sig**2))
    
def gauss_analyt_sol(x,mu,sig,c,t):
    return (gauss_t(x,mu,sig,c,t) + gauss_t(x,mu,sig,c,-t))/2
```

## Introducing a source
The source function, $s(x,t)$, is assumed to be a short pulse which originates in an area where the velocity field is constant.
As a consequence, the source leads to a travelling wave-packet which then scatters from the medium of interest. Such seperation of time-scales allows mapping a time-dependent source to the initial condition of the wave.

The resulting wave packet can be evaluated using the retarded Green's function
$$\psi(x,t) = \int_0^L dx' \int_{-\infty}^{\infty} dt' G(x,x';t-t') s(x',t')~~.$$ 

The Green's function $G(x,x';t-t')$ satisfies the relation 
$$(\partial_t^2 - c^2 \partial_x^2) G(x,x';t-t') = \delta(x-x')\delta(t-t')~~,$$
with $G(\cdot ,t<t') = 0$, $x,x' \in \left[0,L \right]$. Considering periodic boundary conditions $G(x,x';t-t') = G(x+L,x';t-t')$, one obtains
 
The general solution is (d'Alembert / inhomogeneous form) $$u(x,t) = \frac{1}{2c}\int_0^t \int_{x-c(t-\tau)}^{x+c(t-\tau)} s(x',\tau) dx' d\tau$$.
Assuming a point source $s(x,t) = \delta(x-x_0)g(t)$, where $[g] = [u] L/T^2$, one obtains

$$ u(x,t) = \frac{1}{2c} \int_0^{t-|x-x_0|/c} g(\tau)d\tau~~.$$

Since there is a direct relation between the pulse state and the propagating wave packet, we simplify the analysis by considering an initial state of a wave packet propagating at constant speed $c$. It is straight forward to deduce from the integral relation the associated pulse required to create such an initial state. 

The experiment involves a wave packet, initialized on in the vecinity of the left boundary of the grid, propagating towards the scattering medium, which is located around the center of the grid. The medium scatters the wave packet and measurements are performed near the boundaries of the grid.

```python
# Defining functions
def gaussian_wave_packet(x,x0,sig,A):
    '''Generates a gaussian wave packet propagating at a constant speed

        Input:
            x: array, grid
            x0: float, initial gaussian center
            sig: float, standard deviation
            A: float, amplitude
        
        Returns:
            array, concatenated array containing the initial state and velocity
    '''
    N = A/np.sqrt(2*np.pi*sig**2)
    gauss = N*np.exp(-(x-x0)**2/(2*sig**2))
    velocity = ((x-x0)/(sig**2))*gauss
    return np.concatenate((gauss,velocity))
```

## Measurement

```python
def measurement(vec,d_j,Nmax,O,Nt,tup,meas_tup):
    """
    Returns the measurements conducted space time points

    Input: 
        vec: numpy array, containing the initial state
        dj: numpy array, the Chebychev coefficients
        O: function, the dynamical generator (e.g, a differential operator)
        tup: tuple, containing the dynamical generator additional parameters
        meas_tup: list, containing tuples of space time points (time,[position idices]) 
        
    Returns:
            measurements: list, containing tuples of (time, [field at detector positions]) of 
            the measured pressures at the associated space time points 

            vec: array, field at time T = Nt*dt
    """
    meas_inds = [int(meas_tup[i][0]/dt) for i in range(len(meas_tup))]
    measurements = []
    fi = np.zeros((vec.shape[0],3),dtype = complex)
    k = 0
    for j in range(Nt):     # Running over the time steps
        fi[:,0] = vec
        # The normalized differential operator O
        fi[:,1] = O(vec,tup)              
        G1 = d_j[0]*fi[:,0]
        G2 = d_j[1]*fi[:,1]
        G_3 = G1+G2
        for i in range(1,Nmax-1):
            fi[:,2] = 2*O(fi[:,1],tup)-fi[:,0]
            fi[:,0]=fi[:,1]
            fi[:,1]=fi[:,2]
            G_3 = G_3+d_j[i+1]*fi[:,2]
        vec = G_3
        # storing the measurement results in measurements
        if j in meas_inds:
            time = j*dt
            local_meas = []
            # storing the measements at the detector positions at a certain time
            for pos_ind in meas_tup[k][1]:
                local_meas.append(vec[pos_ind])
            k += 1
            measurements.append((time,local_meas))
    return measurements, vec
            
```

## Complete dynamical simulation
In the following we combine all the part to form a complete simulation of the experiment: 
- a source eliminates the medium
- the acoustic wave propagates through the medium
- the wave is detected

```python
# source and initial state
x0 = L/6
sig = L/10
A = 0.1
y0 = gaussian_wave_packet(x,x0,sig,A)

# setting measurement times and positions
counter = range(Nt)
time_interv = Nt//5
time_ind = np.arange(time_interv,Nt+1,time_interv)
meas_time = [dt*ind for ind in time_ind]
pos_ind_1 = Nx//4   
pos_ind_2 = Nx*3//4 
pos_1 = x[pos_ind_1]
pos_2 = x[pos_ind_2]
meas_tup = [(meas_time[i],[pos_ind_1, pos_ind_2]) for i in range(len(meas_time))]



# propagation and recording measurements
tup = (lam_min,lam_max,dk,c)
measurements, final_state  = measurement(y0,d_j,Nmax,O,Nt,tup,meas_tup)
#print(final_state)
```

```python
# propagation check
Nt = 40
x0 = L/6
sig = L/30
A = 0.1
y0 = gaussian_wave_packet(x,x0,sig,A)
y_t = propagator(y0,d_j,Nmax,O,Nt,tup)
u_t = y_t[:Nx]
```
